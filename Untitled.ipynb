```python
import pandas as pd
df = pd.read_csv("churn.csv")
```

```python
#remove/replace special characters and white spaces
df.columns = [i.replace(' ','_').replace("'",'').lower() for i in df.columns]
#Check out how the resultant dataset look like
df.head()
```

```python
def calc_total_mins(day_mins,eve_mins):
    "Output: total_mins"
    total_mins = day_mins + eve_mins
    return total_mins
```
